```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv("Email Analysis Dataset - Email Analysis Dataset.csv")

# Display first five rows
print(df.head())

# Dataset information
print(df.info())
```

```python
# Missing values
print(df.isnull().sum())

# Check duplicate rows
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df["Date"] = pd.to_datetime(df["Date"])



# Extract date information

df["Hour"] = df["Date"].dt.hour

df["Day"] = df["Date"].dt.day_name()

df["Month"] = df["Date"].dt.month_name()
```

```python
print("Total Emails:", len(df))
```

```python
top_sender = df["From Name"].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=top_sender.values,
            y=top_sender.index,
            palette="viridis")

plt.title("Most Active Email Senders")
plt.xlabel("Emails Sent")
plt.ylabel("Employee")
plt.show()
```

```python
top_receiver = df["To Name"].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=top_receiver.values,
            y=top_receiver.index,
            palette="magma")

plt.title("Most Active Email Receivers")
plt.xlabel("Emails Received")
plt.ylabel("Employee")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x="From Department",
              order=df["From Department"].value_counts().index)

plt.xticks(rotation=45)
plt.title("Emails Sent by Department")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x="From seniority",
              order=df["From seniority"].value_counts().index)

plt.title("Emails by Employee Seniority")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df,
              y="Email topic",
              order=df["Email topic"].value_counts().index)

plt.title("Most Common Email Topics")
plt.show()
```

```python
plt.figure(figsize=(5,5))

df["Is opened?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Email Open Rate")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(data=df,
              x="Within workdays",
              palette="Set3")

plt.title("Emails Sent During Workdays")
plt.show()
```

```python
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

plt.figure(figsize=(10,5))

sns.countplot(data=df,
              x="Day",
              order=order)

plt.xticks(rotation=45)
plt.title("Emails Sent by Day")
plt.show()
```

```python
print("\nTop 5 Senders")
print(df["From Name"].value_counts().head())

print("\nTop 5 Receivers")
print(df["To Name"].value_counts().head())

print("\nTop Departments")
print(df["From Department"].value_counts())

print("\nEmail Sentiments")
print(df["Sentiment"].value_counts())

print("\nDevices Used")
print(df["Device"].value_counts())

print("\nWork Hour Emails")
print(df["Within work hours"].value_counts())

print("\nWorkday Emails")
print(df["Within workdays"].value_counts())
```
